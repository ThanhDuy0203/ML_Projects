```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv(r"D:\FPTUniversity\Semester 4\AIL303m\Course 2\boston.csv")#replace path
df
```

```python
X = df[['AGE', 'LSTAT']]
y = df['MEDV']
```

```python
print(X)
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 16, test_size= 0.2)
```

```python
print(X_train)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_transform = scaler.fit_transform(X_train)
```

```python
X_test_transform = scaler.fit_transform(X_test)
```

```python
#Question_1
print(X_train_transform)
```

```python
#Question 4
print(X_train_transform.shape)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
output =LR.fit(X_train_transform, y_train)
```

```python
#Question 5
LR.intercept_
```

```python
#Question 7 ???? 
y_pred = LR.predict(X_test_transform)
print(y_pred[0])
```

```python
#Question 8
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
print('MSE: ', mean_squared_error(y_test, y_pred))
print('MAE: ', mean_absolute_error(y_test, y_pred))
print('MAPE: ', mean_absolute_percentage_error(y_test, y_pred))
```

```python
#Question 9
from sklearn.metrics import r2_score
y_pred_train = LR.predict(X_train_transform)
print('R-Square: ', r2_score(y_train, y_pred_train))
```
